### Enter your Tiled API key

Before running the next cell, make sure you have a valid **Tiled API key** for the server.

The next cell will ask you to enter the key securely, like a password.  
When prompted, paste your API key and press **Enter**.

This will store the key temporarily as an environment variable named:

`TILED_API_KEY`

After entering the key, run the following cells to connect to the Tiled server.

```python
import os
from getpass import getpass

os.environ["TILED_API_KEY"] = getpass("Enter your Tiled API key: ")
```

```python
import os
from tiled.client import from_uri

client = from_uri(
    "http://10.46.217.241:9091",
    api_key=os.environ["TILED_API_KEY"],
)

list(client)
```
