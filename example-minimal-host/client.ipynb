```python
from tiled.client import from_uri

# c = from_uri("http://127.0.0.1:8000", api_key="")
c = from_uri("http://127.0.0.1:8000")

print(list(c))
```

```python
scan = c["scan_a"]
```

```python
image = scan["image"][:]
spectrum = scan["spectrum"][:]

print(image.shape)
print(spectrum.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
plt.colorbar()
plt.title("scan_a image")
plt.show()
```

```python
plt.plot(spectrum)
plt.title("scan_a spectrum")
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```

## Important to see that:

- When you do:

    - scan["image"]

        - This is NOT data yet
        - It is a lazy object

- Only when you do:

    - scan["image"][:]

        - Data is actually fetched from server

- Partial reads (VERY useful later)
    - scan["image"][0:10, 0:10]

        - Only fetch small region
        - critical for large microscopy data
